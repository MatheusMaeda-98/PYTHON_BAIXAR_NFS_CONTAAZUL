```python
from selenium import webdriver
from selenium.webdriver.edge.options import Options as EdgeOptions
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException



# Definir as opções do navegador
edge_options = webdriver.EdgeOptions()
prefs = {
    "download.default_directory": r"C:\Users\mathy\Documentos",
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
}
edge_options.add_experimental_option("prefs", prefs)

# Inicializar o WebDriver do Microsoft Edge com o serviço e as opções
driver = webdriver.Edge(options=edge_options)


# Inicializar o WebDriver

import time

import re

# Abra a página de login do Conta Azul
driver.get("https://login.contaazul.com/#/")

# Aguarde alguns segundos para a página carregar completamente (você pode ajustar esse tempo conforme necessário)
time.sleep(15)


# Verifique se o botão de aceitar cookies está disponível
# Espere até que o botão "Aceitar cookies" seja visível
try:
    cookie_button = WebDriverWait(driver, 20).until(
        EC.visibility_of_element_located((By.XPATH, "/html/body/div[2]/div[2]/a[1]"))
    )
    cookie_button.click()
    print("Botão 'Aceitar cookies' clicado com sucesso.")
except:
    print("Botão 'Aceitar cookies' não encontrado ou não clicável. Continuando sem aceitar cookies.")



# Aguarde alguns segundos para a página carregar completamente (você pode ajustar esse tempo conforme necessário)
driver.implicitly_wait(20)

# Localize o campo de entrada de e-mail e insira o endereço de e-mail
try:
    email_input = driver.find_element(By.XPATH, "/html/body/div[4]/div/div[1]/div/div/div[2]/div/div/div[3]/div/div/form/div/div/div[1]/div/div/div[1]/input")
    email_input.send_keys("'your email'")
    print("Email preenchido com sucesso.")
except:
    print("Campo de e-mail não encontrado.")


driver.implicitly_wait(20)

# Localize o campo de senha e insira a senha
try:
    password_input = driver.find_element(By.XPATH, "/html/body/div[4]/div/div[1]/div/div/div[2]/div/div/div[3]/div/div/form/div/div/div[2]/div/div/div[1]/input")
    password_input.send_keys("'your password'")
    print("Senha preenchida com sucesso.")
except:
    print("Campo de senha não encontrado.")

# Envie o formulário de login (pressionando a tecla Enter)
password_input.send_keys(Keys.ENTER)

# Aguarde alguns segundos para que o login seja processado (você pode ajustar esse tempo conforme necessário)
time.sleep(30)

driver.get("https://app.contaazul.com/#/ca/financeiro/contas-a-pagar")

time.sleep(30)
#___________________________________________________________________________________________________

# Logado no ContaAzul

#clicar o periodo do mês

periodo_button = WebDriverWait(driver, 200).until(
EC.visibility_of_element_located((By.XPATH, "/html/body/caf-application-layout/div/caf-application/div/div[2]/div[2]/div/main/div/div[2]/div/section/div[3]/div[2]/div[2]/div/div/div/div/div[2]/div/div[1]/div[2]/div[1]/div/div/div/div/div/div[1]/span/button"))
)
periodo_button.click()
    
#clicar o periodo personalizado

periodo_personalizado_button = WebDriverWait(driver, 100).until(
EC.visibility_of_element_located((By.XPATH, "/html/body/caf-application-layout/div/caf-application/div/div[2]/div[2]/div/main/div/div[2]/div/section/div[3]/div[2]/div[2]/div/div/div/div/div[2]/div/div[1]/div[2]/div[1]/div/div/div/div/div/div[2]/div[8]")))
periodo_personalizado_button.click()

#________________________________________________________________________________________________

#Clicar data de inicio

data_inicio_button = WebDriverWait(driver, 40).until(
EC.visibility_of_element_located((By.XPATH, "/html/body/div[15]/form/div/div/div/div/div[1]/span/div/div[1]/input")))
data_inicio_button.click()

for _ in range(5):
    # Simule a pressão da tecla Backspace
    data_inicio_button.send_keys(Keys.BACKSPACE)
    time.sleep(1)
    
    data_inicio_button.send_keys(Keys.DELETE)
    time.sleep(1)
    
    data_inicio_button.send_keys(Keys.DELETE)
    time.sleep(1)

    
# Agora vamos preencher a data de inicios

data = "01042023"  # Número que você deseja digitar, será a data inicial sem as barras (ddmmaa)
time.sleep(3)
data_inicio_button.send_keys(data)

#____________________________________________________________________________________________________________

#Clicar data final

data_final_button = WebDriverWait(driver, 100).until(
EC.visibility_of_element_located((By.XPATH, "/html/body/div[15]/form/div/div/div/div/div[2]/span/div/div[1]/input"))
)
data_final_button.click()

#vai primeiro apagar a data nela
for _ in range(5):
    # Simule a pressão da tecla Backspace
    data_final_button.send_keys(Keys.BACKSPACE)
    time.sleep(1)  # Aguarde 1 segundo entre as pressões de tecla (opcional)
    
    # Simule a pressão da tecla Delete
    data_final_button.send_keys(Keys.DELETE)
    time.sleep(1)  # Aguarde 1 segundo entre as pressões de tecla (opcional)
    
# Agora vamos preencher a data de inicios

data2 = "30042023"  # Número que você deseja digitar, será a data inicial sem as barras (ddmmaa)
data_final_button.send_keys(data2)

#_____________________________________________________________________________________________________________

#Clicar o botão aplicar

aplicar_button = WebDriverWait(driver, 30).until(
EC.visibility_of_element_located((By.XPATH, "/html/body/div[15]/form/div/div/footer/div/span/button/div"))
)
aplicar_button.click()

#______________________________________________________________________________________________________________

# APERTAR BOTÃO PESQUISAR
PESQUI_button = WebDriverWait(driver, 70).until(
EC.visibility_of_element_located((By.XPATH, "/html/body/caf-application-layout/div/caf-application/div/div[2]/div[2]/div/main/div/div[2]/div/section/div[3]/div[2]/div[2]/div/div/div/div/div[2]/div/div[1]/div[2]/div[2]/div/div/form/div/div/div/div[1]/input"))
)
PESQUI_button.click()

PESQU = "PSIK"  # Número que você deseja digitar, será a data inicial sem as barras (ddmmaa)

PESQUI_button.send_keys(PESQU)

PESQUI_button.send_keys(Keys.ENTER)

time.sleep(15)

#_______________________________________________________________________________________________________________

# vai repetir 12 vezes pois vai aplicar esse processo em 12 meses

for _ in range (1): 
    
    time.sleep(2)
    select_buttons1 = driver.find_element(By.XPATH, "/html/body/caf-application-layout/div/caf-application/div/div[2]/div[2]/div/main/div/div[2]/div/section/div[3]/div[2]/div[2]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/table/thead/tr/th[1]/div/div/div/div/input")
    select_buttons1.click()
    time.sleep(2)
    

    # Extrair o código HTML da página atual
    html_code = driver.page_source

    # Parsear o código HTML
    soup = BeautifulSoup(html_code, 'html.parser')

    # Encontrar o elemento span com a classe desejada
    span_element = soup.find('span', class_='ds-batch-table__top-bar__counter')

    # Se o elemento span for encontrado
    if span_element:
        # Extrair o texto do elemento
        texto = span_element.get_text(strip=True)

        # Extrair o número usando expressões regulares
        match = re.search(r'(\d+)', texto)

        if match:
            numero_registros = int(match.group(1))
            print("Número de registros:", numero_registros)
            # Agora você pode usar o número de registros onde for necessário
        else:
            print("Número de registros não encontrado.")
    else:
        print("Elemento span não encontrado.")

    
    
# Agora, encontre todos os botões de download

    
    for i in range(1, numero_registros + 1):

        time.sleep(3)

    # Encontrar o número total de elementos tr
        xpath = "/html/body/caf-application-layout/div/caf-application/div/div[2]/div[2]/div/main/div/div[2]/div/section/div[3]/div[2]/div[2]/div/div/div/div/div[2]/div/div[2]/div/div[1]/div/div[2]/div/table/tbody/tr[" + str(i) + "]/td[5]"

        # Localize o elemento usando o XPath construído
        element = driver.find_element(By.XPATH, xpath)

        # Faça algo com o elemento (por exemplo, clicar nele)
        element.click()
        # Faça algo com o elemento td[5] (por exemplo, clicar nele)
        time.sleep(3)


        for x in range(1, 5):  

            dw_elements = "/html/body/div[15]/div[2]/div[2]/div/div[1]/div/div[3]/div/div/div[2]/div/div/div/div/div/div/div/table/tbody/div/tr[" + str(x) + "]/td[1]/span/a/b"

            # Loop para iterar sobre os elementos tr

                 # Encontrar o elemento td[5] dentro do tr atual
            try:
                # Procurar o elemento td[5] dentro do tr atual
                dw_5 = driver.find_element(By.XPATH, dw_elements)
                time.sleep(2)
                dw_5.click()
                time.sleep(3)

            except NoSuchElementException:
                print("Elemento não encontrado para o elemento:", x)
                # Você pode adicionar mais tratamento de erro aqui, se necessário
                continue  # Continuar para a próxima iteração do loop
                time.sleep(2)

        #botao de sair
        time.sleep(2)
        download_buttons1 = driver.find_element(By.XPATH, "/html/body/div[15]/div[2]/div[1]/div/div/span/button")
        download_buttons1.click()
        time.sleep(2)
        
# Mudar o mês pra frente
    time.sleep(2)
    avance_buttons = driver.find_element(By.XPATH, "/html/body/caf-application-layout/div/caf-application/div/div[2]/div[2]/div/main/div/div[2]/div/section/div[3]/div[2]/div[2]/div/div/div/div/div[2]/div/div[1]/div[2]/div[1]/div/div/button[2]")
    avance_buttons.click()
    time.sleep(3)    
                                   
driver.Quit()
```
